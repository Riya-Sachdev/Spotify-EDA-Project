## Null Handling

```python
print("Total Null Values in our Data Set :", df.isnull().sum().sum())
print("Total Null Values in our Data Set (column wise):\n\n")
print(df.isnull().sum())
```

```python
df = df.dropna(subset=["artist_name"])
```

```python
df["artist_genres"] = df["artist_genres"].fillna("Unknown")
```

```python
print("Total Null Values in our Data Set after fillna :", df.isnull().sum())
```

#### Insights
All columns are complete after filling missing values, except **album_release_date**, which has **8,579 missing entries**. This is perfectly fine to have null values **(NaT - Not a Time)** in this column, as some albums may not have a **recorded release date**.
