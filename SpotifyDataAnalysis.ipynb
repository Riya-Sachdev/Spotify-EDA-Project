### 6. Relationships Between Numerical Features

#### Track Duration vs Track Popularity

```python
plt.figure(figsize = (5,4))
sbn.scatterplot(x = "track_duration_min" , y = "track_popularity" , data = df)
plt.title("Track Duration vs Track Popularity")
plt.show()
```

## Insights

The scatter plot illustrates the relationship between *track duration (in minutes)* and *track popularity*.

### Key Observations
- Most tracks fall between *2 to 5 minutes*, indicating this is the most common song length.
- Popularity scores are *widely spread* across all durations, ranging from very low to very high.
- The *highest concentration of popular tracks (60–90 popularity)* appears within the *3–4 minute* range.
- Tracks longer than *8 minutes* are relatively rare and tend to have *lower or more inconsistent popularity*.
- Very short tracks (under 2 minutes) generally do not reach high popularity scores.

### Overall Trend
There is *no strong linear correlation* between track duration and popularity. However, the plot suggests that *moderately timed tracks (around 3–4 minutes)* are more likely to achieve higher popularity.

### Conclusion
Track duration alone does not determine popularity, but songs with a *standard commercial length* tend to perform better. This implies that other factors—such as artist popularity, genre, marketing, and listener preferences—play a more significant role in determining a track’s success.

#### Artist Followers vs Artist Popularity

```python
plt.figure(figsize = (5,4))
sbn.scatterplot(x = "artist_followers" , y = "artist_popularity" , data = df)
plt.title("Artist Followers vs Artist Popularity")
plt.show()
```

## Insights

The scatter plot shows the relationship between an artist’s *number of followers* and their *artist popularity score*.

### Key Observations
- There is a *strong positive relationship* between artist followers and artist popularity.
- Artists with *very low follower counts* show a wide range of popularity scores, from very low to moderately high.
- As the number of followers increases, popularity scores tend to *increase and stabilize* between *80–95*.
- Artists with *tens of millions of followers* almost exclusively fall into the *high popularity range*.
- The curve appears *non-linear*, with popularity increasing rapidly at first and then leveling off.

### Overall Trend
Popularity grows quickly with early gains in followers but shows *diminishing returns* at higher follower counts. After a certain point, additional followers do not significantly increase the popularity score.

### Conclusion
Artist popularity is *strongly influenced by follower count*, especially at lower levels. However, once an artist reaches a large audience, popularity tends to plateau, suggesting Spotify’s popularity metric may cap or normalize scores for top artists.

### 8. Release Trends Over Time

#### Extract Release Year

```python
df.loc[:,"release_year"] = df.loc[:,"album_release_date"].dt.year
df.loc[:,"release_year"]
```

```python
yearly_tracks = df.groupby("release_year").size()

plt.figure(figsize = (5,4))
yearly_tracks.plot()
plt.title("Number of Tracks Released Over Time")
plt.xlabel("Year")
plt.ylabel("Track Count")
plt.show()
```

A line chart was used to analyze the trend in the number of tracks released per year.

### Observations:

* From 1950 to the late 1980s, track releases remain very low and relatively stable.

* A gradual increase begins in the early 1990s, indicating growth in music production.

* A sharp rise is observed after 2010, with track releases increasing rapidly.

* The most recent years show the highest number of releases, highlighting a strong upward trend.

### Insight:
The significant increase in track releases in recent years reflects the impact of digital streaming platforms, easier access to music production tools, and the growth of independent artists. The dataset is heavily concentrated in recent years, which should be considered when interpreting popularity and trend analyses.

### 9. Correlation Analysis Using Heat Map

```python
# heatmap --> correlation
corr_data = df[num_variables].corr()
```

```python
plt.figure(figsize = (6,5))
sbn.heatmap(
      corr_data,
      annot = True,
      cmap = "coolwarm")
plt.title("Correlation Using Heat Map")
plt.show()
```

A correlation heatmap was used to examine the relationships between key numerical variables in the dataset.

### Observations:

* Track popularity shows a moderate positive correlation with artist popularity (0.47), indicating that tracks by more popular artists tend to perform better.

* Artist popularity has a strong positive correlation with artist followers (0.64), suggesting that artists with larger fan bases are generally more popular.

* Track popularity has a weak positive correlation with artist followers (0.23), implying follower count alone does not strongly determine a track’s success.

* Album total tracks shows little to no correlation with track popularity (−0.05), indicating album size does not influence individual track performance.

* Track duration has a negligible correlation with track popularity (0.11), suggesting song length does not significantly impact popularity.

### Insight:
The analysis indicates that artist-related factors (popularity and followers) have a greater influence on track popularity than track-specific attributes such as duration or album size. This highlights the importance of an artist’s existing audience in driving streaming success.

### 11. Combined Artist Ranking (Overall Performance)

```python
artist_summary = df.groupby("artist_name").agg({
    "track_popularity" : "mean",
    "artist_popularity" : "mean",
    "artist_followers" : "mean"
})
artist_summary["combined_score"] = (
    artist_summary["track_popularity"] * 0.4 +
    artist_summary["artist_popularity"] * 0.3 +
    artist_summary["artist_followers"] / artist_summary["artist_followers"].max() * 0.3
)
artist_summary.sort_values("combined_score", ascending =  False).head(5)
```
