### Statistical Distribution of Numerical Variables 

```python
num_variables = [
    "track_popularity",
    "artist_popularity",
    "artist_followers",
    "album_total_tracks",
    "track_duration_min"
]
df[num_variables].describe()
```

### Insights

#### Popularity
- **Track popularity**: Median **58**, IQR **39–71** → wide spread from unpopular to very popular tracks.
- **Artist popularity**: Generally high (median **74**), but ranges from **0–100**.

#### Artists
- **Followers**: Highly skewed.
  - Median: **~6.1M**
  - Mean: **~24M**
  - Max: **145M**
  → A few mega-artists heavily raise the average.

#### Albums 
- **Album size**: Typical album has **13 tracks** (IQR **6–17**), but some extreme outliers (**up to 181 tracks**).+

#### Track Duration
- **Track length**: Fairly consistent.
  - Median: **3.45 min**
  - Most tracks fall between **2.9–4.0 min**
  - Few long outliers (max **13.5 min**).

#### Key Takeaways
- Popularity and followers are **right-skewed** with strong outliers.
- Track duration is **stable**, unlike popularity and album size.
- Dataset represents a mix of mainstream and niche artists.

### Histograms and Density Plots

```python
x = ["track_popularity", "artist_popularity", "artist_followers", "album_total_tracks", "track_duration_min"]
for i in x:
    plt.figure(figsize = (6,3))
    plt.title(f"Distribution of {i}")
    sbn.histplot(x = i , data = df , bins = 24 , kde = True)
    plt.show()
```

### Key Findings

- **Track Popularity**
  - Distribution is right-skewed with most tracks having popularity scores between 40 and 80.
  - A notable spike at zero popularity indicates many less-known or new tracks.

- **Artist Popularity**
  - Concentrated at higher popularity levels (60–90), reflecting well-known artists in the dataset.
  - Artist popularity is generally higher and less variable compared to track popularity.

- **Artist Followers**
  - Highly skewed distribution with most artists having few followers.
  - A few artists have extremely large followings, representing significant outliers.

- **Album Total Tracks**
  - Most albums contain between 5 and 20 tracks.
  - Some albums have a large number of tracks, likely compilations or deluxe editions.

- **Track Duration**
  - Normally distributed with a mean around 3 to 4 minutes.
  - Typical of commercial music, with few very short or very long tracks.

### Insights
- Popularity and follower features are skewed and may benefit from transformations for modeling.
- Track duration is stable and can be treated as a reliable feature.
- The dataset includes a mix of mainstream popular artists and many lesser-known ones, offering diverse analysis opportunities.
